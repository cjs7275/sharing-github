I Based my Entity Relationship diagram off of six key entities for the pet shop. These entities were for the customers, food, pets, enclosures, transactions and the loyalty program. Each box includes attributes of each entity such as the customers first and last name or the type of food. Each entity has a relationship(verb) with another, A customer will buy food or adopt a pet and in return they will get a record of their transaction. Pets are kept in their own enclosures and food is fed to the pets. Customers also have the option to enroll in the loyalty program. The diagram also shows CArdinality with its connection lines. For example the customers to the pets are many to many but the customers to the loyalty program is many to one because there are potentially many customers but only one loyalty program.
